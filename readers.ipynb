### OpenCV

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Leer la original_image
original_image = cv2.imread('../images/x.png')

# Convertir la original_image a escala de grises
image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)

# # Mostrar la imagen en pantalla
# cv2.imshow('Imagen en Escala de Grises', image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# # Guardar la imagen en escala de grises
# cv2.imwrite('imagen_gris.jpg', image)

plt.imshow(image, cmap='gray')
plt.axis('off')
plt.tight_layout()
plt.show()

print(image)
```

### Pillow

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Abrir la imagen
imagen = Image.open('../images/x.png')

# Convertir la imagen a escala de grises
image = imagen.convert('L')

# # Mostrar la imagen en pantalla
# image.show()

# # Guardar la imagen en escala de grises
# image.save('imagen_gris.jpg')

# Mostrar la imagen en escala de grises debajo de la celda
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.tight_layout()
plt.show()

image_matrix = np.array(image)
print(image_matrix)
```
